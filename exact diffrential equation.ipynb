```python
# Python code for Exact Diffrential equation
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, integrate
```

### Question1 : Solve  (3(x**2)*(y**3) â€“ 5*(x**4))dx + (y + 3*(x**3)* (y**2))dy = 0

```python
# Define symbols
x, y,c = symbols('x y c')

# Assign value of M and N
M = (3*(x**2)*(y**3))-5*(x**4)
N = (y + 3*(x**3)*(y**2))

# Check for exact diffrential equation or not
dM_dy = M.diff(y)
dN_dx = N.diff(x)
# print("dM_dy:" , dM_dy)
# print("dN_dx:", dN_dx)

if M.diff(y) == N.diff(x):
    print("It's an exact differential equation.")
else:
    print("It's not an exact differential equation. Find an integrating factor")
```

```python
I1 = integrate(M, x)
# print(I1)
I2 = integrate(N.subs(x, 0), y)
# print(I2)
general_solution = I1 + I2 + c
print("Solution is: ", general_solution,"where c is arbitrary constant")
```

### Question2 : Solve (x**2+3*(y**2))dx + 2*x*ydy = 0

```python
# Define symbols
x, y,c = symbols('x y c')

# Assign value of M and N
M = ((x**2)+ (3*(y**2)))
N = (2*x*y)

# Check for exact diffrential equation or not
dM_dy = M.diff(y)
dN_dx = N.diff(x)
# print("dM_dy:" , dM_dy)
# print("dN_dx:", dN_dx)

if M.diff(y) == N.diff(x):
    print("It's an exact differential equation.")
else:
    print("It's not an exact differential equation. Find an integrating factor")
```

```python
from sympy import simplify, exp
expression1 = simplify((dM_dy - dN_dx) / N)
# print(expression1)
expression2 = simplify((-dM_dy + dN_dx) / M)
if expression1.has(x) and not expression1.has(y):
    IF = exp(integrate(expression1, x))
#     print(IF)
elif expression2.has(y) and not expression2.has(x):
    IF = exp(integrate(expression2, y))   
else :
    print('Use hit and trail to calculate IF')
```

```python
M1 = M * IF
N1 = N * IF
# print(M1,N1)
I1 = integrate(M1, x)
# print(I1)
I2 = integrate(N1.subs(x, 0), y)
# print(I2)
general_solution = I1 + I2 + c
print("Solution is: ", general_solution,"where c is arbitrary constant")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
sampling_rate = 1000  # Hz, sampling rate
duration = 1  # seconds
num_samples = int(sampling_rate * duration)
frequency_1 = 10  # Hz
frequency_2 = 50  # Hz

# Generate time vector
time = np.linspace(0, duration, num_samples, endpoint=False)

# Generate synthetic signal (spectrum)
signal = np.sin(2 * np.pi * frequency_1 * time) + 0.5 * np.sin(2 * np.pi * frequency_2 * time)

# Plot the synthetic signal
plt.figure(figsize=(10, 4))
plt.plot(time, signal)
plt.title('Synthetic Signal (Spectrum)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Perform Fourier Transform
frequency_values = np.fft.fftfreq(num_samples, 1/sampling_rate)  # Frequency values
fft_values = np.fft.fft(signal)  # Fourier Transform

# Plot the Fourier Transform
plt.figure(figsize=(10, 4))
plt.plot(frequency_values, np.abs(fft_values))
plt.title('Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
