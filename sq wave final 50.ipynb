```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.signal import square

from scipy.integrate import quad

from math import* #import all function from math

x=np.arange(-np.pi,np.pi,0.001) #x axis has been chosen from –π to +π 

y=square(x) #defining square wave function 𝑦 =−1, 𝑓𝑜𝑟 − 𝜋 ≤ 𝑥 ≤ 0


#define fuction

fc=lambda x:square(x)*cos(i*x)  #i :dummy index

fs=lambda x:square(x)*sin(i*x)

n=50 #max value of I, not taken infinity, better result with high value

An=[] # defining array

Bn=[]

sum=0
# color_cycle = plt.cm.jet(np.linspace(0, 1, 20))

for i in range(n):

     an=quad(fc,-np.pi,np.pi)[0]*(1.0/np.pi)

     An.append(an)

for i in range(n):

     bn=quad(fs,-np.pi,np.pi)[0]*(1.0/np.pi)

     Bn.append(bn) #putting value in array Bn

for i in range(n):

    if i==0.0:

         sum=sum+An[i]/2

    else:

         sum=sum+(An[i]*np.cos(i*x)+Bn[i]*np.sin(i*x))
#          plt.plot(x,sum,'g')
#          plt.plot(x, sum, '-', color=color_cycle[i - 1], linewidth=1.2, label=f'n={i+1}')
#          plt.legend(loc='lower right', fontsize='small')
         
            
#     plt.plot(x,sum,'g')

plt.plot(x,sum,'g')

plt.plot(x,y,'r--')
plt.plot(x, y, 'r--', label='Original Square Wave')
plt.title("Fourier series for square wave")

plt.show()
```
